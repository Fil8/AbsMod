```python
# some useful functions:

def readFile(parameterFile):
        
    parameters={}
    
    try:
        paramFile=open(parameterFile)
    except:
        print "%s file not found"  % parameterFile
        return 1
        
    paramList=paramFile.readlines()
    paramFile.close()     

    
    for line in paramList:
        if line.strip():  # non-empty line?
            tmp=line.split('=')
            tmp2=tmp[0].split('[')
            key, value = tmp2[0],tmp[-1]  # None means 'all whitespace', the default
            parameters[key] = value
    
    return parameters

def ra2deg(rad):
    
        ra=string.split(rad,':')
    
        hh=float(ra[0])*15
        mm=(float(ra[1])/60)*15
        ss=(float(ra[2])/3600)*15
        
        return hh+mm+ss

def dec2deg(decd):
        dec=string.split(decd,':')
        
        hh=abs(float(dec[0]))
        mm=float(dec[1])/60
        ss=float(dec[2])/3600
        return hh+mm+ss
    
def ang2lin(z,dl,ang): # r in arcsec
    
    #dl = dl/3.085678e24 # Mpc
    r = ang * dl / (RAD2DEG * 3600 * (1+z)**2) # Mpc
    
    return r

def lin2ang(z,dl,r): # r in Mpc

    ang = RAD2DEG * 3600. * r * (1.+z)**2 / dl # arcsec

    return ang
```

```python
##### upload parameter file to create the disk 

rootdir='/Users/maccagni/Documents/notebooks/Mod_Abs/AbsMod_git/'
root_obs='../observations/3c305/'

#READ the parameter file
#uncomment if you set parameter file from terminal 
#file=sys.argv[1]
fileinput = rootdir+'../par_3c305.txt'
#set a dictionary with the parameters -> call readFile function
par=readFile(fileinput)

#define the 
#filecont=sys.argv[2]
filecont = root_obs+'3c305.fits'

#root_out=rootdir+'3c293/'
#filespec=sys.argv[3]
filespec = root_obs+'spec_305.txt' #spectrum in Hz



RAD2DEG=180./math.pi
HI_hz=1.42040575177e9
C=2.99792458E5 

#set a dictionary with the parameters -> call readFile function
par=readFile(fileinput)
#set the variables
RMAX=float(par.get('rmax'))
RMIN=float(par.get('rmin'))
H_0=float(par.get('h0'))
I=float(par.get('i'))
PA=float(par.get('pa'))


RMAXD2=float(par.get('rmax_in'))
RMIND2=float(par.get('rmin_in'))
H_0D2=float(par.get('h0_in'))
ID2=float(par.get('i_in'))
PAD2=float(par.get('pa_in'))

VROT=float(par.get('vrot'))

SIGN=float(par.get('sign'))

FLUX_VALUE=float(par.get('flux'))
D_L=float(par.get('d_l'))
z_red=float(par.get('z'))
RA=par.get('ra')
DEC=par.get('dec')
RES=float(par.get('pix_res'))
VRES=float(par.get('vel_res'))
DISP=float(par.get('disp'))
PA_C=float(par.get('pa_cont'))
CONT_LIM=float(par.get('flux_cont_lim'))
VSYS=float(par.get('v_sys'))


ra=ra2deg(RA)
dec=dec2deg(DEC)

#define the trigonometric parameters of the disk
I_rad=math.radians(I)
PA_rad=math.radians(PA)
I_radD2=math.radians(ID2)
PA_radD2=math.radians(PAD2)
PA_C_rad=math.radians(PA_C)


cos_I = np.cos(I_rad)
sin_I = np.sin(I_rad)

cos_PA = np.sin(PA_rad)
sin_PA = np.cos(PA_rad)



print 'Variables set\n'
for keys,values in par.items():
    print keys+' = '+str(values)
    
    
    
```

```python
vels=np.arange(-VROT*2,VROT*2+VRES,VRES)

print vels
print 'Cube set'

#DEFINE THE GAUSSIAN FOR THE CONVOLUTION
arg=-((vels*vels)/(2*DISP*DISP))
gauss=1./(np.sqrt(2*np.pi)*DISP)*np.exp(arg)
```

```python
#Functions for the disk

def flux(r):

    rho=FLUX_VALUE*np.ones(r.shape)
    
    flux=+rho
    
    return flux    

def space(x_cube,y_cube,z_cube,continuum_cube):

        #allocate the output array
        #transformation into disk coordinates
        x=cos_PA*x_cube+sin_PA*y_cube
        
        
        y=cos_I*(-sin_PA*x_cube+cos_PA*y_cube)+sin_I*z_cube
        
        
        z=-sin_I*(-sin_PA*x_cube+cos_PA*y_cube)+cos_I*z_cube
        
        
        #determine the radius of the disk
        
        r=np.sqrt(np.power(x,2)+np.power(y,2))

        angle=np.arctan2(y,x)
        
        
        vel=-SIGN*sin_I*np.cos(angle)*VROT
        
        flusso = flux(r)
            
        #condition I am in the disk
        #if r <=RMAX and abs(z)<=H_0/2.:    
        #set to zero - -999 the flux and velocities outside the disk
        idx = (r > RMAX) & (abs(z)>=H_0/2.) & (r<RMIN)
        
        #flusso[idx] = 0.0
        continuum_cube[idx] = 0.0

        index_disk = flusso < 1.0
        vel[index_disk]= -999.


        #set to non good values what is behind the continuum
        index_abs = (z_cube>np.tan(PA_C_rad)*x_cube)

        vel[index_abs] = -999.
        #flusso[index_abs] = 0.0
        
        continuum_cube[index_abs] = 0.0
        
        
        return vel,continuum_cube
    
    
def build_continuum(x_los,y_los):
        
        
        #load continuum and define the centre
        f=pyfits.open(filecont)
        dati=f[0].data
        head=f[0].header
        dati=np.squeeze(dati)
        dati=np.squeeze(dati)

        print dati.shape

        scale_cont_asec=head['CDELT2']*3600
        scale_cont_pc=ang2lin(z_red,D_L,scale_cont_asec)*1e6
        print 'Continuum pixel size [pc]'
        print scale_cont_pc

        head=fits.getheader(filecont)

        del head['CTYPE4']
        del head['CDELT4']    
        del head['CRVAL4']
        del head['CRPIX4']
        del head['CRPIX3'] 
        del head['CRVAL3']
        del head['CDELT3']
        del head['CTYPE3']
        del head['NAXIS3']
        del head['NAXIS4']        
        del head['NAXIS']
        del head['CROTA1']
        del head['CROTA2']
        del head['CROTA3']
        del head['CROTA4']


        w=wcs.WCS(head)    

        #cen_x,cen_y=w.wcs_world2pix(ra,dec,0)
        cen_x,cen_y=w.wcs_world2pix(ra,dec,1)

        print '**** Continuum centre ****' 
        print cen_x,cen_y
        print '\n'
        
        print x_los[-1]
        
        x_los_num_right=x_los[-1]/scale_cont_pc
        x_los_num_left=x_los[0]/scale_cont_pc
        
        y_los_num_up=y_los[-1]/scale_cont_pc
        y_los_num_low=y_los[0]/scale_cont_pc
        
        
        y_up=cen_y+y_los_num_up
        x_right=cen_x+x_los_num_right
        
        y_low=cen_y+y_los_num_low
        x_left=cen_x+x_los_num_left
        

        #continuum value linear interpolation
        
        x_left_int=math.modf(x_left)
        x_right_int=math.modf(x_right)
        
        y_low_int=math.modf(y_low)
        
        y_up_int=math.modf(y_up)
        
        
        sub_dati=dati[int(y_low_int[1]):int(y_up_int[1]),int(x_left_int[1]):int(x_right_int[1])]
        
        #determine how much I have to interpolate     
        zoom_factor= float(len(x_los))/float(len(sub_dati[0]))
        zoom_dati=zoom(sub_dati,zoom_factor,order=1)
        
        
        #mask the noise of the continuum:
        index_mask = sub_dati < CONT_LIM
        
        sub_dati[index_mask] = 0.0
        
        
                
        return zoom_dati
```

Here the program begins: 

- i cut through the cube fixing y coordinate and moving in the horizontal direction (x), then going to the next line. 
- i cut through the cube (z direction) from $-2\times R_{\rm MAX}$ to $2\times R_{\rm MAX}$
- for each line of sight I half the resolution and check if the spectra satisfy the condition
- the condition is given by the maximum absolute difference

```python
#MAIN MAIN MAIN

#set timer
tempoinit=time.time()

print 'Beginning'

#initial conditions on the output integrated spectrum 
spec_int=np.zeros([2,len(vels)])
spec_int[0,:]=vels[:]


#build the cube which contains my disk based on its dimensions
#and on the specified resolution
x_los=np.arange(-RMAX*1.2,+RMAX*1.2+RES,RES)
y_los=np.arange(-RMAX*1.2,+RMAX*1.2+RES,RES)
z_los=np.arange(-RMAX*1.2,+RMAX*1.2+RES,RES)


#load the continuum image interpolated at the specified resolution
continuum_image=build_continuum(x_los,y_los)


#create a meshgrid
x_cube,y_cube,z_cube = np.meshgrid(x_los,y_los,z_los)

#create a cube containing the fluxes of the continuum image
continuum_cube=np.dstack([continuum_image]*x_cube.shape[0])




velocity,flusso = space(x_cube,y_cube,z_cube,continuum_cube)

#interpolate to X time the resolution
vel_zoom = zoom(velocity, 2,order=3)
flux_zoom = zoom(flusso, 2, order=3)

vel_index= (vel_zoom >= -VROT) & (vel_zoom <= VROT) 
lin_vel=vel_zoom[vel_index]
lin_flux=flux_zoom[vel_index]


for i in xrange(0,len(vels)-1):
    
    #print vels[i]
    #print vels[i+1]
    index=(vels[i]<lin_vel) & (lin_vel < vels[i+1])
    #print indexpr
    #print f[index]
    
    spec_int[1,i]=-np.sum(lin_flux[index])
    
#shift output spectrum to make it symmetric with respect to vel = 0 [binning problem]

#specint[1,:]=-specint[1,:]
#specint=[vel_bins,flux_bins]


print 'end \n'

tempofin=(time.time()-tempoinit)/60.
tempoinit=tempofin
print "total time %f minutess" % (tempofin)  
  
print 'NORMAL TERMINATION'


```
